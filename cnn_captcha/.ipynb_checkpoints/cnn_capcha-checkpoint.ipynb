## 导入配置

```python
from captcha.image import ImageCaptcha
import matplotlib.pyplot as plt
import numpy as np
import random
from PIL import Image
import random
import string
import os
import pickle
from keras.layers import * 
from keras.models import Model
from keras_preprocessing import image
from keras.utils import plot_model
from keras.callbacks import *
from tensorflow import keras
import pydot
import graphviz
import pydot_ng
%matplotlib inline 
%config InlineBackend.figure_format = 'retina' 
```

## 测试验证码生成

```python
characters = string.digits
print(characters)

width, height, n_len, n_class = 170, 80, 4, len(characters)
print(n_class)
generator = ImageCaptcha(width=width, height=height)
random_str = ''.join([random.choice(characters) for j in range(n_len)])
img = generator.generate_image(random_str)

plt.imshow(img)
plt.title(random_str)
```

## 定义验证码生成器

```python
def gen(batch_size=32):
    X = np.zeros((batch_size, height, width, 3), dtype=np.uint8)
    y = [np.zeros((batch_size, n_class), dtype=np.uint8) for i in range(n_len)]
    generator = ImageCaptcha(width=width, height=height)
    while True:
        for i in range(batch_size):
            random_str = ''.join([random.choice(characters) for j in range(4)])
            X[i] = generator.generate_image(random_str)
            for j, ch in enumerate(random_str):
                y[j][i, :] = 0
                y[j][i, characters.find(ch)] = 1
        yield X, y
```

## One hot解码

```python
def decode(y):
    y = np.argmax(np.array(y), axis=2)[:,0]
    return ''.join([characters[x] for x in y])
```

## 进行测试

```python
X, y = next(gen(1))
plt.imshow(X[0])
plt.title(decode(y))
```

## 建立模型
利用吴恩达的那个HappyModel作业,仿照建立一个神经网络模型, 这里结构仿照的是vgg16, 适用于分类问题

```python
def CapchaModel(input_shape):
    """
    Implementation of the CapchaModel.
    
    Arguments:
    input_shape -- shape of the images of the dataset

    Returns:
    model -- a Model() instance in Keras
    """
    
    ### START CODE HERE ###
    # Feel free to use the suggested outline in the text above to get started, and run through the whole
    # exercise (including the later portions of this notebook) once. The come back also try out other
    # network architectures as well. 
    X_input = Input(input_shape)
    X=X_input
    for i in range(5):
        X = Conv2D(32 * 2 ** i, (3, 3), activation='relu', padding='same')(X)
        X = BatchNormalization(epsilon=1e-05,momentum=0.1)(X)
        X = MaxPool2D((2, 2))(X)
    X=Flatten()(X)
    X=Dropout(0.25)(X)
    X = [Dense(n_class, activation='softmax', name='c%d'%(i+1))(X) for i in range(4)]
    model = Model(inputs = X_input, outputs = X, name='CapchaModel')
    ### END CODE HERE ###
    
    return model
```

```python
capchaModel = CapchaModel((height, width,3))
```

```python
capchaModel.summary()
```

```python
plot_model(capchaModel,show_shapes=True,to_file='model.png')
```

```python
capchaModel.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

# 加载模型并预测

```python
capchaModel = CapchaModel((height, width,3))
capchaModel.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
capchaModel.load_weights("./cnn_best.h5")
```

```python
from tqdm import tqdm
def evaluate(model, batch_num=32):
    batch_acc = 0
    generator = gen()
    for i in tqdm(range(batch_num)):
        X, y = next(generator)
        y_pred = capchaModel.predict(X)
        y_pred = np.argmax(y_pred, axis=2).T
        y_true = np.argmax(y, axis=2).T
        temp = np.mean(list(map(np.array_equal, y_true, y_pred)))
        print(temp)
        batch_acc += temp
    return batch_acc / batch_num
```

```python
evaluate(capchaModel)
```

```python
X, y = next(gen(1))
y_pred = capchaModel.predict(X)
plt.title('real: %s\npred:%s'%(decode(y), decode(y_pred)))
plt.imshow(X[0], cmap='gray')
```

## 再训练,利用模型在训练

```python
callbacks = [EarlyStopping(patience=3), CSVLogger('cnn.csv'), ModelCheckpoint('cnn_best.h5', save_best_only=True)]
history=capchaModel.fit_generator(gen(),
                                      validation_data=gen(), steps_per_epoch=100,
                                      epochs=5, validation_steps=10
                                      )
capchaModel.save('./captch_model-1.h5')
capchaModel.load_weights('./captch_model-1.h5')
```
